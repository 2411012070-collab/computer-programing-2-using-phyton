```python
from transaksi.diskon import hitung_diskon
from transaksi.pajak import hitung_ppn

harga_awal = 1000000
diskon = 10

potongan = hitung_diskon(harga_awal, diskon)
harga_setelah_diskon = harga_awal - potongan
ppn = hitung_ppn(harga_setelah_diskon)
total = harga_setelah_diskon + ppn

print("Harga Awal              :", harga_awal)
print("Potongan Diskon         :", int(potongan))
print("Harga Setelah Diskon    :", int(harga_setelah_diskon))
print("PPN 11%                 :", int(ppn))
print("Total Bayar             :", int(total))
```
